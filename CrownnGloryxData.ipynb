```python
# Group total_by_payment by payment options
grouped = total_by_payment.groupby('Payment_Method')['Total_Transaction'].sum().reset_index()
# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(grouped['Payment_Method'], grouped['Total_Transaction'])
plt.xlabel('Payment Method')
plt.ylabel('Total Transaction Amount')
plt.title('Total Transaction Amounts by Payment Method')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
# This code will create  a bar chart showing 
#the total transaction amounts for each payment method







```

```python
# Calculate the total sales for each payment method
total_sales = grouped['Total_Transaction'].sum()

# Calculate the percentage of total sales for each payment method
grouped['Percentage'] = (grouped['Total_Transaction'] / total_sales) * 100

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(grouped['Percentage'], labels=grouped['Payment_Method'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Percentage of Total Sales by Payment Method')
plt.show()
# This code will create a pie chart showing the percentage 
# distribution of total sales among different payment methods
```
