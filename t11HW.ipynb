### 5.b - Distribution plot (challenging!)

Plot the distributions of fare paid for survivors and deceased in a way that makes for a good visual comparison.

```python
fare_live = titanic[titanic['Survived']==1]
fare_dead = titanic[titanic['Survived']==0]

# Create a figure with two subplots
fig, axs = plt.subplots(ncols=2, figsize=(12, 5))

# Plot the distribution of fares paid by survivors
sns.histplot(fare_live['Fare'], ax=axs[0], kde=True, color='blue')
axs[0].set_title('Fare Distribution for Survivors')

# Plot the distribution of fares paid by those who died
sns.histplot(fare_dead['Fare'], ax=axs[1], kde=True, color='red')
axs[1].set_title('Fare Distribution for Non-Survivors')

# Show the plot
plt.show()

```
